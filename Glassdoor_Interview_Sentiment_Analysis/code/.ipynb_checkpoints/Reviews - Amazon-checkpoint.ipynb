```python
# URLs
url = 'https://www.glassdoor.sg/Reviews/Amazon-Reviews-E6036.htm' # 52k reviews total, scrap 1500 pages
```

```python
next_url = None
page = 1
num = 0
main_dict = []
prevent = [37,73]
max_page = 1501             # Change the page number to the number of pages you want to web scrape
start_time = time.time()

print('Scraping page :')

while page < max_page:
    if next_url == None:
        next_url = url    

    else:
        next_url = 'https://www.glassdoor.com'+after

    if page % prevent[num] == 0:
        print(f'Pause to rest. Run time : {round((time.time() - start_time)/60,2)} minutes\n')
        time.sleep(random.randint(3,5))
        print('Scraping page :')
        num = 1 - num

    driver.get(next_url)
    time.sleep(random.randint(3,4))

    post = html.fromstring(driver.page_source)
    get_id = post.xpath('//*[contains(@id,"empReview_")]')[0:]

    while len(get_id) < 10:
        print(f'Refreshing page {page}...')
        driver.get(next_url)
        time.sleep(random.randint(5,8))
        post = html.fromstring(driver.page_source)
        get_id = post.xpath('//*[contains(@id,"empReview_")]')[0:]

    for i in range(len(get_id)):

        id_pos = get_id[i].attrib['id']
        comment_ = post.xpath(f'//*[@id="{id_pos}"]/div/div[2]/div[2]/div[1]/h2/a/text()')
        employment_ = post.xpath(f'//*[@id="{id_pos}"]/div/div[2]/div[2]/div[1]/p/text()')
        pro_ = post.xpath(f'//*[@id="{id_pos}"]/div/div[2]/div[2]/div[2]/div[1]/p[2]/span/text()')
        con_ = post.xpath(f'//*[@id="{id_pos}"]/div/div[2]/div[2]/div[2]/div[2]/p[2]/span/text()')
        rating_ = post.xpath(f'//*[@id="{id_pos}"]/div/div[2]/div[2]/div[1]/div[1]/span/div/div/div/text()')

        page_frame = {}

        if len(comment_) == 1:
            page_frame['comment'] = comment_[0]

        if len(employment_) == 1:
            page_frame['employment'] = employment_[0]

        if len(pro_) == 1:
            page_frame['pro'] = pro_[0]

        if len(con_) == 1:
            page_frame['con'] = con_[0]

        if len(rating_) == 1:
            page_frame['rating'] = rating_[0]

        main_dict.append(page_frame)

    if page % 50 == 0:
        print(f'{page} of {max_page-1}, Run time : {round((time.time() - start_time)/60,2)} minutes')

    else:
        print(f'{page} of {max_page-1}')

    after1 = post.xpath('//*[@id="NodeReplace"]/main/div/div[1]/div[6]/ul/li[7]/a/@href')
    after2 = post.xpath('//*[@id="NodeReplace"]/main/div/div[1]/div[7]/ul/li[7]/a/@href')

    if len(after1) == 1:
        after = after1[0]

    else:
        after = after2[0]


    amazon_df = pd.DataFrame(main_dict)                                        # Change the data frame and csv name
    amazon_df.to_csv('../datasets/amazon_reviews.csv', index = False)
    page+=1

print(f'\nSuccessfully scraped for Amazon reviews.\nTotal duration : {round((time.time() - start_time)/60,2)} minutes')
# Change the name of company you are scraping
```
