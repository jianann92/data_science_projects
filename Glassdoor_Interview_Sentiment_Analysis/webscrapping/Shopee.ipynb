```python
url = 'https://www.glassdoor.sg/Interview/Shopee-Interview-Questions-E1263091.htm' # Change URL to company you want
next_url = None
page = 1
num = 0
main_dict = []
prevent = [17,37]
max_page = 68            # Change the page number to the number of pages you want to web scrape
start_time = time.time()

print('Scraping page :')

while page < max_page:
    if next_url == None:
        next_url = url    
    
    else:
        next_url = 'https://www.glassdoor.com'+after
    
    if page % prevent[num] == 0:
        print(f'Pause to rest. Run time : {round((time.time() - start_time)/60,2)} minutes\n')
        time.sleep(random.randint(3,5))
        print('Scraping page :')
        num = 1 - num
    
    driver.get(next_url)
    time.sleep(random.randint(1,3))
    
    post = html.fromstring(driver.page_source)
    get_id = post.xpath('//*[contains(@id,"InterviewReview")]')[1:]
 
    while len(get_id) < 10:
        print(f'Refreshing page {page}...')
        driver.get(next_url)
        time.sleep(random.randint(2,4))
        post = html.fromstring(driver.page_source)
        get_id = post.xpath('//*[contains(@id,"InterviewReview")]')[1:]

    for i in range(len(get_id)):
        
        id_pos = get_id[i].attrib['id']
        interview_ = post.xpath(f'//*[@id="{id_pos}"]/div[3]/div/div[2]/div[2]/div/div[1]/p[4]/text()')
        question_ = post.xpath(f'//*[@id="{id_pos}"]/div[3]/div/div[2]/div[2]/div/div[2]/ul/li/span/text()')
        offer_ = post.xpath(f'//*[@id="{id_pos}"]/div[3]/div/div[2]/div[1]/div/div[1]/div/div[2]/span/text()')
        experience_ = post.xpath(f'//*[@id="{id_pos}"]/div[3]/div/div[2]/div[1]/div/div[2]/div/div[2]/span/text()')
        difficulty_ = post.xpath(f'//*[@id="{id_pos}"]/div[3]/div/div[2]/div[1]/div/div[3]/div/div[2]/span/text()')
        
        page_frame = {}
        
        if len(interview_) == 1:
            page_frame['interview'] = interview_[0]
            
        if len(question_) == 1:
            page_frame['question'] = question_[0]
        
        if len(offer_) == 1:
            page_frame['offer'] = offer_[0]

        if len(experience_) == 1:
            page_frame['experience'] = experience_[0]

        if len(difficulty_) == 1:
            page_frame['difficulty'] = difficulty_[0]
        
        main_dict.append(page_frame)

    if page % 50 == 0:
        print(f'{page} of {max_page-1}, Run time : {round((time.time() - start_time)/60,2)} minutes')
        
    else:
        print(f'{page} of {max_page-1}')
    
    page+=1
    after = post.xpath('//*[@id="FooterPageNav"]/div[2]/ul/li[7]/a/@href')[0]
    shopee_df = pd.DataFrame(main_dict)                                        # Change the data frame and csv name

print(f'\nSuccessfully scraped for Shopee interviews.\nTotal duration : {round((time.time() - start_time)/60,2)} minutes')
# Change the name of company you are scraping
```

```python
# processing text, without lemmatizing or stemming
p_stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
# stop_set = ['the', 'a', 'an', 'i', 'he', 'she', 'they', 'to', 'of', 'it', 'from']
stop_words = set(stopwords.words('english'))
stop_words |= set(['interview','interviewer','question','recruiter','process','ask','amazon','phone','questions',
                   'apple','facebook','google','microsoft','round', 'one'])
stop_words.remove('not')

# Adding CSS class icon into list of words to remove
swear_words_corpus = requests.get("http://www.cs.cmu.edu/~biglou/resources/bad-words.txt").text # Swear words to be removed
swear_words = re.sub('[\n]', ' ', swear_words_corpus)
emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags=re.UNICODE)


corpus = []
# def processing_text(texts): # Function to process texts
for i in range(0,len(df)):
    
    word = emoji_pattern.sub(r'', df['interview'][i]) # no emoji
    word = BeautifulSoup(word,'lxml').getText()
    word = re.sub('[^a-zA-Z]',' ',word.lower())    # Remove non-letters and convert all to lower case
    word = word.split() # Split into individual words
    
    clean = [w for w in word if w not in swear_words]
    clean = [w for w in word if w not in stop_words] # Removing stopwords and stemming remaining words
#     clean = [lemmatizer.lemmatize(w) for w in word if w not in stop_words] # Removing stopwords and lemmatizing remaining
    clean = ' '.join(clean)
    corpus.append(clean)
```

```python
cleaned_corpus = []

for i in corpus:
    word = i.split() # Split into individual words
#     clean = [p_stemmer.stem(w) for w in word] # Removing stopwords and stemming remaining words
    clean = [lemmatizer.lemmatize(w) for w in word] # Removing stopwords and lemmatizing remaining
    clean = [w for w in clean if w not in stop_words]
    clean = [w for w in clean if len(w) >= 2] # Remove all words that has single alphabets
    clean = ' '.join(clean)
    cleaned_corpus.append(clean)
```

```python
df['text'] = cleaned_corpus
```
